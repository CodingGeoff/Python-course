```python
import os
import shutil

def move_files_safely():
    # 1. å®šä¹‰ç›®æ ‡æ–‡ä»¶å¤¹åç§°
    target_folder = "assets"
    
    # 2. å¦‚æœæ–‡ä»¶å¤¹ä¸å­˜åœ¨ï¼Œåˆ™åˆ›å»º
    if not os.path.exists(target_folder):
        os.makedirs(target_folder)
        print(f"ğŸ“‚ å·²åˆ›å»ºæ–‡ä»¶å¤¹: {target_folder}")
    
    # 3. è·å–å½“å‰ç›®å½•ä¸‹æ‰€æœ‰ .xlsx å’Œ .docx æ–‡ä»¶
    # ä½¿ç”¨ lower() ç¡®ä¿ .XLSX æˆ– .DOCX ä¹Ÿèƒ½è¢«è¯†åˆ«
    files_to_move = [f for f in os.listdir('.') 
                     if os.path.isfile(f) and f.lower().endswith(('.xlsx', '.docx', '.csv'))]
    
    if not files_to_move:
        print("âœ… å½“å‰ç›®å½•æ²¡æœ‰éœ€è¦ç§»åŠ¨çš„æ–‡ä»¶ã€‚")
        return

    print(f"ğŸ” å‘ç° {len(files_to_move)} ä¸ªæ–‡ä»¶ï¼Œå‡†å¤‡ç§»åŠ¨...")
    
    success_count = 0
    fail_count = 0

    # 4. éå†å¹¶ç§»åŠ¨æ–‡ä»¶
    for filename in files_to_move:
        src_path = filename
        dst_path = os.path.join(target_folder, filename)
        
        # --- æ ¸å¿ƒé€»è¾‘ï¼šå†²çªå¤„ç† (é‡å‘½å) ---
        # å¦‚æœç›®æ ‡è·¯å¾„å·²ç»å­˜åœ¨æ–‡ä»¶ï¼Œåˆ™æ·»åŠ åç¼€ _1, _2, _3...
        base_name, extension = os.path.splitext(filename)
        counter = 1
        
        while os.path.exists(dst_path):
            new_name = f"{base_name}_{counter}{extension}"
            dst_path = os.path.join(target_folder, new_name)
            counter += 1
        
        # --- å®‰å…¨ç§»åŠ¨ ---
        try:
            shutil.move(src_path, dst_path)
            print(f"  âœ… ç§»åŠ¨æˆåŠŸ: {filename} -> {dst_path}")
            success_count += 1
        except Exception as e:
            # å¦‚æœæŠ¥é”™ï¼ˆä¾‹å¦‚æ–‡ä»¶è¢«å ç”¨ï¼‰ï¼Œæ‰“å°é”™è¯¯ä½†ä¸åœæ­¢ç¨‹åº
            print(f"  âŒ ç§»åŠ¨å¤±è´¥: {filename}")
            print(f"     é”™è¯¯ä¿¡æ¯: {e}")
            fail_count += 1

    print("-" * 30)
    print(f"ğŸ‰ å¤„ç†å®Œæˆï¼æˆåŠŸ: {success_count} ä¸ªï¼Œå¤±è´¥: {fail_count} ä¸ª")

# æ‰§è¡Œå‡½æ•°
move_files_safely()
```
