```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 设置pandas显示选项，取消输出省略
pd.set_option('display.max_rows', None)  # 显示所有行
pd.set_option('display.max_columns', None)  # 显示所有列
pd.set_option('display.width', None)  # 自动适配宽度
pd.set_option('display.max_colwidth', None)  # 显示列的完整内容

# 模拟复杂数据
np.random.seed(42)
dates = pd.date_range('20250101', periods=100)
df = pd.DataFrame({
    'Date': dates,
    'Category': np.random.choice(['A', 'B', 'C'], 100),
    'Value_1': np.random.randn(100).cumsum(), # 随机游走
    'Value_2': np.random.rand(100) * 100,
    'Status': np.random.choice(['Active', 'Inactive'], 100, p=[0.7, 0.3])
})

# 故意引入缺失值演示清洗
df.loc[5:10, 'Value_1'] = np.nan

print(df)

```
