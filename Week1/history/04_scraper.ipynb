```python
# HTTP è¯·æ±‚é…ç½®
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# è®¾ç½®pandasæ˜¾ç¤ºé€‰é¡¹ï¼Œå–æ¶ˆè¾“å‡ºçœç•¥
pd.set_option('display.max_rows', None)  # æ˜¾ç¤ºæ‰€æœ‰è¡Œ
pd.set_option('display.max_columns', None)  # æ˜¾ç¤ºæ‰€æœ‰åˆ—
pd.set_option('display.width', None)  # è‡ªåŠ¨é€‚é…å®½åº¦
pd.set_option('display.max_colwidth', None)  # æ˜¾ç¤ºåˆ—çš„å®Œæ•´å†…å®¹

# çˆ¬å–ç›®æ ‡: Python.org
URL = "https://www.python.org"

# 1. è®¾ç½®ä¼ªè£… Headers (æ¨¡æ‹ŸçœŸå®æµè§ˆå™¨)
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.9'
}

try:
    start_time = time.time()
    response = requests.get(URL, headers=headers, timeout=10)
    response.raise_for_status() # æ£€æŸ¥ 404/500 é”™è¯¯
    print(f"âœ… è¯·æ±‚æˆåŠŸ! çŠ¶æ€ç : {response.status_code}")
    print(f"â±ï¸ è€—æ—¶: {time.time() - start_time:.4f} ç§’")
except requests.exceptions.RequestException as e:
    print(f"âŒ è¯·æ±‚å¤±è´¥: {e}")

```

```python
# BeautifulSoup è§£æä¸ CSS é€‰æ‹©å™¨
soup = BeautifulSoup(response.text, 'html.parser')

# 1. æå–æ‰€æœ‰å¯¼èˆªé“¾æ¥ (ä½¿ç”¨ CSS Selector)
# æŸ¥æ‰¾ nav æ ‡ç­¾ä¸‹çš„ li ä¸‹çš„ a æ ‡ç­¾
nav_links = soup.select('nav li a') 

data_list = []
for link in nav_links:
    text = link.get_text(strip=True)
    href = link.get('href')
    # å¤„ç†ç›¸å¯¹è·¯å¾„
    if href and href.startswith('/'):
        href = URL + href
    if text:
        data_list.append({'Category': 'Nav', 'Text': text, 'URL': href})

# 2. æå– Latest News (å‡è®¾ç»“æ„)
news_items = soup.select('.blog-widget li')
for item in news_items:
    try:
        date = item.find('time').get_text(strip=True)
        title = item.find('a').get_text(strip=True)
        data_list.append({'Category': 'News', 'Text': title, 'Date': date})
    except AttributeError:
        continue

```

```python
# æ•°æ®ç»“æ„åŒ–ä¸å¯¼å‡º
df = pd.DataFrame(data_list)
print(f"\n=== æŠ“å–ç»“æœ ({len(df)} æ¡) ===")
display(df.head(10))

# è¿‡æ»¤ç‰¹å®šæ•°æ®
news_df = df[df['Category'] == 'News']
print("\n=== æœ€æ–°æ–°é—» ===")
display(news_df)

```
